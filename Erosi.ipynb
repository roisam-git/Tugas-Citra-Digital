```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Daftar path gambar
paths = [
    "C:/Users/hp/Documents/tugascitradigital/1.png",
    "C:/Users/hp/Documents/tugascitradigital/2.png"
]

# Siapkan figure 2 baris  2 kolom (asli & erosi untuk setiap gambar)
plt.figure(figsize=(10, 8))

for i, path in enumerate(paths):
    # Baca gambar dalam mode grayscale
    img = cv2.imread(path, 0)

    if img is None:
        print(f" Gambar ke-{i+1} tidak ditemukan: {path}")
        continue

    # Threshold biner
    _, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

    # Kernel erosi
    kernel = np.ones((3, 3), np.uint8)

    # Proses erosi
    eroded = cv2.erode(binary, kernel, iterations=1)

    # Simpan hasil ke file
    output_path = f"C:/Users/hp/Documents/tugascitradigital/eroded_{i+1}.png"
    cv2.imwrite(output_path, eroded)

    # Plot gambar asli
    plt.subplot(2, 2, i*2 + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f"Gambar Asli {i+1}")
    plt.axis("off")

    # Plot hasil erosi
    plt.subplot(2, 2, i*2 + 2)
    plt.imshow(eroded, cmap='gray')
    plt.title(f"Hasil Erosi {i+1}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
